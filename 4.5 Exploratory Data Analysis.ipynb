```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("housing.csv")
```

```python
df.head()
```

 I used the housing dataset because this dataset contains information about different houses and their prices. The dataset includes features like income, median age of the house, number of bedrooms, population, and households.
The dataset has many rows, which makes it good for analyzing the patterns.

```python
df.columns
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x="median_income", y="median_house_value")
plt.title("Median Income vs House Prices")
plt.show()
```

This scatter plot shows the relationship between income and house price.
There is a strong positive relationship. As income increases, house value also increases.

```python
plt.figure(figsize=(13, 11))
sns.scatterplot(data=df, x="housing_median_age", y="median_house_value")
plt.title("Housing Median Age vs Median House Value")
plt.show()
```

This shows how the age of a house relates to its price.
There isnt a strong relationship because House value does not change much based on the age.

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x="total_rooms", y="median_house_value")
plt.title("Total Rooms vs House Value")
plt.show()
```

This shows how the number of rooms relates to price.
There is some relationship, but it is not very strong compared to income.

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x="population", y="median_house_value")
plt.title("Population vs House Value")
plt.show()
```

This graph compares population and house prices.
There is no clear pattern really, so population does not strongly affect house value.

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x="households", y="median_house_value")
plt.title("Households vs Median House Value")
plt.show()
```

This graph shows households compared to prices.
There is a weak relationship, but not very strong.

Which Features Are Most Related?
The feature that is most related to house value is median income, because the graph shows a clear upward trend.

Which Features Are Not Strongly Related?
Features like population and housing_median_age do not show a strong relationship with house value.
