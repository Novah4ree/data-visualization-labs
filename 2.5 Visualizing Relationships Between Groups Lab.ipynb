```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
df = pd.read_csv(r"loan_data.csv")
df.head()
```

```python
df.columns
```

Age vs Loan Status

This boxplot shows the relationship between a person's age and whether their loan was approved or rejected.
Looking at the chart, the age ranges for approved and rejected applicants overlap a lot. The medians are also pretty close. This suggests that age does not seem to have a strong effect on whether a loan gets approved.
Based on this visualization, age does not appear to be a major factor the model uses when making its decision.

```python
plt.figure(figsize=(6,7))

sb.boxplot(
    data=df,
    x="loan_status",
    y="person_age",
)

plt.title("Relationship Between Age and Loan Approval")
plt.xlabel("Loan Status (F = Rejected, A = Approved)")
plt.ylabel("Applicant Age")

plt.show()
```

Income vs Loan Status

This visualization compares a person's annual income with loan approval.
From the boxplot, approved applicants generally have higher incomes compared to rejected applicants. The median income for approved loans is higher.
This means income likely plays a role in the model's decision because higher income equals a better ability to repay the loans.

```python
plt.figure(figsize=(8,7))

sb.boxplot(
    data=df,
    x="loan_status",
    y="person_income"
)

plt.title("Applicant Income vs Loan Status")
plt.xlabel("Loan Approval Status")
plt.ylabel("Applicant Annual Income")

plt.show()
```

Interest Rate vs Loan Status

This boxplot compares loan interest rate with loan approval status.
The distributions overlap quite a bit, which suggests interest rate itself may not be the main factor used to approve or reject a loan. Instead, the interest rate is likely determined after the applicant's risk level is evaluated.

```python
plt.figure(figsize=(10,6))

sb.boxplot(
    data=df,
    x="loan_status",
    y="loan_int_rate"
)

plt.title("Interest Rate vs Loan Approval Status")
plt.xlabel("Loan Approval Status")
plt.ylabel("Loan Interest Rate")

plt.show()
```

Loan Percent of Income vs Loan Status

This plot shows how large the loan is compared to the applicant's income.
Applicants who requested loans that were a smaller percentage of their income were more likely to get approved. When the loan becomes a larger than their income, less likely to get approved.
This because lenders want to make sure borrowers can afford the loan.

```python
plt.figure(figsize=(10,6))

sb.boxplot(
    data=df,
    x="loan_status",
    y="loan_percent_income"
)

plt.title("Loan Burden vs Loan Approval")
plt.xlabel("Loan Approval Status")
plt.ylabel("Loan as % of Income")

plt.show()
```

Credit Score vs Loan Status

This boxplot shows how credit score relates to loan approval.
There is a noticeable difference between the two groups. Applicants with approved loans tend to have higher credit scores overall.
Credit score is one of the strongest factors influencing whether a loan gets approved.

```python
plt.figure(figsize=(9,6))

sb.boxplot(
    data=df,
    x="loan_status",
    y="credit_score"
)

plt.title("Credit vs Loans")
plt.xlabel("Loan Status Approval")
plt.ylabel("Credit Score Approval")

plt.show()
```

Credit History Length vs Loan Status

This shows credit history and whether their loan was approved.
Applicants with longer credit histories are more likely to be approved. A longer good credit history gives lenders more information about a person's past borrowing behavior.

```python
plt.figure(figsize=(10,6))

sb.boxplot(
    data=df,
    x="loan_status",
    y="cb_person_cred_hist_length"
)

plt.title("Credit History Length vs Approval Status")
plt.xlabel("Loan Approval Status")
plt.ylabel("Length of Credit History")

plt.show()
```
